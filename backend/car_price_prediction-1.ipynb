*****Import  required packages*****

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("vehicles.csv")
df.head(50)
```

```python
# Clean cylinders column
df['cylinders'] = df['cylinders'].str.replace(' cylinders', '').str.replace('cylinder', '',regex=False)
```

```python
# # Handle missing values
# df = df.dropna(subset=['price', 'year', 'manufacturer', 'model', 'odometer'])
```

```python
# Convert posting_date to datetime and extract day/month
df['posting_date'] = pd.to_datetime(df['posting_date'],utc=True)
df['posting_day'] = df['posting_date'].dt.day
df['posting_month'] = df['posting_date'].dt.month
df.head()
```

```python
#columns to keep
cols_to_keep = ['price', 'year', 'manufacturer', 'model', 'condition', 
               'cylinders', 'fuel', 'odometer', 'title_status', 'transmission',
               'drive', 'type', 'paint_color', 'state', 'posting_day','posting_month','lat','long','region']
df = df[cols_to_keep]
```

```python
#replacing numeric null values by taking median
num_cols= df.select_dtypes(include=['int64','float64',]).columns
df[num_cols]=df[num_cols].fillna(df[num_cols].median())

#replacinng categorical null values by taking mode
cat_cols= df.select_dtypes(include=['object']).columns 
for col in cat_cols :
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
float_cols = ['posting_day','posting_month','year']
df[float_cols] = df[float_cols].astype(int)
```

```python
df=df.drop_duplicates()
```

```python
df.info()
```

```python
# Remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers(df, 'price')
df = remove_outliers(df, 'odometer')
df = remove_outliers(df, 'year')
```

```python
# Calculate age and average mileage per year
current_year = 2025
df['age'] = current_year - df['year']
df['avg_mil'] = df['odometer'] / df['age']
```

```python
# Classify manufacturers by country
manufacturer_mapping = {
    'American': ['harley-davidson', 'chevrolet', 'pontiac', 'ram', 'ford', 'gmc', 
                 'tesla', 'jeep', 'dodge', 'cadillac', 'chrysler', 'lincoln', 
                 'buick', 'saturn', 'mercury'],
    'Japanese': ['lexus', 'nissan', 'toyota', 'acura', 'honda', 'infiniti', 
                 'subaru', 'mitsubishi', 'datsun', 'mazda'],
    'German': ['volkswagen', 'mercedes-benz', 'bmw', 'audi', 'porsche'],
    'Italian': ['ferrari', 'fiat', 'alfa-romeo'],
    'Korean': ['kia', 'hyundai'],
    'Swedish': ['volvo'],
    'English': ['rover', 'mini', 'land rover', 'jaguar']
}

def classify_manufacturer(mfg):
    mfg = str(mfg).lower()
    for country, brands in manufacturer_mapping.items():
        if mfg in brands:
            return country
    return 'Unknown'

df['manufacturer_country'] = df['manufacturer'].apply(classify_manufacturer)
```

```python

# Classify car type
def classify_type(car_type):
    if pd.isna(car_type):
        return 'unknown'
    car_type = car_type.lower()
    
    luxury_small = ['sedan', 'convertible', 'coupe', 'hatchback', 'other']
    luxury_large = ['suv', 'wagon']
    non_luxury_small = ['pickup', 'truck', 'offroad']
    non_luxury_large = ['van', 'mini-van', 'bus']
    
    if car_type in luxury_small:
        return 'luxury_small'
    elif car_type in luxury_large:
        return 'luxury_large'
    elif car_type in non_luxury_small:
        return 'non-luxury_small'
    elif car_type in non_luxury_large:
        return 'non-luxury_large'
    else:
        return 'unknown'

df['type_class'] = df['type'].apply(classify_type)
```

```python
# Classify paint color
def classify_color(color):
    if pd.isna(color):
        return 'unknown'
    color = color.lower()
    
    light_colors = ['white', 'silver', 'yellow', 'orange', 'green', 'custom']
    dark_colors = ['black', 'red', 'blue', 'purple', 'grey', 'brown']
    
    if color in light_colors:
        return 'light'
    elif color in dark_colors:
        return 'dark'
    else:
        return 'unknown'

df['color_class'] = df['paint_color'].apply(classify_color)
```

```python
from sklearn.model_selection import train_test_split
# Select features (X) and target (y)
X = df.drop(['price','lat','long','posting_day','posting_month'],axis=1)
y = df['price']


# Simple train-test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True,random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
def create_lg_model():
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(X_train,y_train)
    return model
```

```python
def create_xgboost_model():
    from xgboost import XGBRegressor
    model = XGBRegressor()
    model.fit(X_train, y_train)
    return model
```

```python
def create_rf_model():
    from sklearn.ensemble import RandomForestRegressor
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    return model
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

def evaluate_model(name,model):
    y_prediction=model.predict(X_test)

    print(f"_ _{name} model evaluation _ _")
    print(f"mean_squared_error:{mean_squared_error(y_test,y_prediction)}")
    print(f"Mean_absolute_error:{mean_absolute_error(y_test,y_prediction)}")
    print(f"r2_score:{r2_score(y_test,y_prediction)}")
    
```

```python
model_lg = create_lg_model()
evaluate_model('linear Regresion',model_lg)
```

```python
model_xgb = create_xgboost_model()
evaluate_model('XGBoost',model_xgb)
```

```python
model_rf = create_rf_model()
evaluate_model('Random Forest',model_rf)
```

```python
# since RandomForest Model  Gives the best r2_score we select the Random forest model. 
```
